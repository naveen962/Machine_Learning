```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('weatherHistory.csv')
```

```python
X=dataset['Humidity']
Y=dataset['Apparent Temperature (C)']
plt.scatter(X,Y,s=5)
plt.show()
```

```python
#Building the model
m=0
c=0

L=0.01

epochs=50000
n=float(len(X))

for i in range(epochs):
    Y_pred=m*X + c
    D_m=(-2/n) * np.sum(X*(Y-Y_pred))
    D_c=(-2/n) * np.sum(Y-Y_pred)
    m=m-L*D_m
    c=c-L*D_c
print(m,c)
```

```python
#making predictions
Y_pred=m*X + c

plt.scatter(X,Y,s=8,color='green')
plt.plot(X,Y_pred,color='red')    
plt.show()
```
