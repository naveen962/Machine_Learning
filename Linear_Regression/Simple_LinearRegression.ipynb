```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
dataset_train=pd.read_csv('train.csv')
dataset_test=pd.read_csv('test.csv')
```

```python
dataset_train.count()
```

```python
dataset_train.dropna(axis=0,inplace=True)
```

```python
X=dataset_train.iloc[:,0]
Y=dataset_train.iloc[:,1]
plt.scatter(X,Y,s=5)
plt.show()
```

```python
#Building the model
m=0
c=0

L=0.0001

epochs=1000
n=float(len(X))

for i in range(len(X)):
    Y_pred=m*X + c
    D_m=(-2/n) * sum(X*(Y-Y_pred))
    D_c=(-2/n) * sum(Y-Y_pred)
    m=m-L*D_m
    c=c-L*D_c
print(m,c)
```

```python
#making predictions
Y_pred=m*X + c

plt.scatter(X,Y,s=8,color='green')
plt.plot(X,Y_pred,color='red')
```

```python
X_test=dataset_test.iloc[:,0]
Y_test=dataset_test.iloc[:,1]
```

```python
Y_pred_test=m*X_test + c

plt.scatter(X_test,Y_test,s=8,color='black')
plt.plot(X_test,Y_pred_test,color='red')
```
